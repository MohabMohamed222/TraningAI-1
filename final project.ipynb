```python
# Download sample CamVid dataset for segmentation/classification
!git clone https://github.com/alexgkendall/SegNet-Tutorial.git

# Images are inside: SegNet-Tutorial/CamVid/
# - images: SegNet-Tutorial/CamVid/train/
# - labels: SegNet-Tutorial/CamVid/trainannot/
import os
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
import tensorflow as tf
from tensorflow.keras.utils import to_categorical
from tensorflow.keras.preprocessing.image import img_to_array, load_img
from tensorflow.keras.models import Model
from tensorflow.keras.layers import *
from tensorflow.keras.optimizers import Adam, RMSprop, SGD
from tensorflow.keras.callbacks import EarlyStopping, ReduceLROnPlateau
from tensorflow.keras.applications import EfficientNetB0
import pandas as pd

# Configure TensorFlow
tf.config.run_functions_eagerly(True)
tf.compat.v1.enable_eager_execution()

# Set memory growth for GPU if available
gpus = tf.config.experimental.list_physical_devices('GPU')
if gpus:
    try:
        for gpu in gpus:
            tf.config.experimental.set_memory_growth(gpu, True)
    except RuntimeError as e:
        print(e)

# 1. Load the dataset with error handling
def load_dataset(image_dir, mask_dir, img_size=(128, 128)):
    images = []
    masks = []
    try:
        # Only consider common filenames in images and masks
        image_files = sorted([f for f in os.listdir(image_dir) if f.endswith(('.png', '.jpg'))])
        mask_files = sorted([f for f in os.listdir(mask_dir) if f.endswith('.png')])
        common_files = [f for f in image_files if os.path.splitext(f)[0] in [os.path.splitext(m)[0] for m in mask_files]]
        if not common_files:
            raise Exception("No matching image/mask pairs found.")
        for fname in common_files:
            img_path = os.path.join(image_dir, fname)
            mask_path = os.path.join(mask_dir, os.path.splitext(fname)[0] + '.png')
            if not os.path.exists(img_path) or not os.path.exists(mask_path):
                continue
            img = load_img(img_path, target_size=img_size)
            mask = load_img(mask_path, target_size=img_size, color_mode="grayscale")
            images.append(img_to_array(img))
            masks.append(img_to_array(mask))
        if len(images) == 0:
            raise Exception("No images loaded. Check your image and mask directories.")
        print(f"Dataset loaded successfully. Total samples: {len(images)}")
        return np.array(images), np.array(masks)
    except Exception as e:
        print("Error loading dataset:", e)
        return None, None

# 2. Split the dataset into train and test sets
def split_dataset(images, masks, test_size=0.2, random_state=42):
    try:
        X_train, X_test, y_train, y_test = train_test_split(
            images, masks, test_size=test_size, random_state=random_state)
        print(f"Dataset split: {X_train.shape[0]} train / {X_test.shape[0]} test samples")
        return X_train, X_test, y_train, y_test
    except Exception as e:
        print("Error splitting dataset:", e)
        return None, None, None, None

# 3. Preprocessing: Rescale images and convert masks to categorical
def preprocess_data(images, masks, num_classes=12):
    try:
        images = images.astype(np.float32) / 255.0  # Normalize images
        masks = masks.squeeze(-1).astype(np.uint8)

        # Ensure mask values are within valid range
        masks = np.clip(masks, 0, num_classes-1)
        masks = to_categorical(masks, num_classes=num_classes)

        print(f"Preprocessing done. Images shape: {images.shape}, Masks shape: {masks.shape}")
        return images, masks
    except Exception as e:
        print("Error during preprocessing:", e)
        return None, None

# 4. Display some sample images and masks
def plot_samples(images, masks, num_samples=3):
    plt.figure(figsize=(10, num_samples * 3))
    for i in range(num_samples):
        idx = np.random.randint(0, images.shape[0])
        plt.subplot(num_samples, 2, 2*i+1)
        plt.imshow(images[idx])
        plt.title('Image')
        plt.axis('off')
        plt.subplot(num_samples, 2, 2*i+2)
        plt.imshow(np.argmax(masks[idx], axis=-1), cmap='tab20')
        plt.title('Mask')
        plt.axis('off')
    plt.tight_layout()
    plt.show()

# 5. Build U-Net Model
def build_unet(input_shape=(128, 128, 3), num_classes=12):
    inputs = Input(input_shape)

    # Encoder
    c1 = Conv2D(64, (3, 3), activation='relu', kernel_initializer='he_normal', padding='same')(inputs)
    c1 = Conv2D(64, (3, 3), activation='relu', kernel_initializer='he_normal', padding='same')(c1)
    p1 = MaxPooling2D((2, 2))(c1)

    c2 = Conv2D(128, (3, 3), activation='relu', kernel_initializer='he_normal', padding='same')(p1)
    c2 = Conv2D(128, (3, 3), activation='relu', kernel_initializer='he_normal', padding='same')(c2)
    p2 = MaxPooling2D((2, 2))(c2)

    c3 = Conv2D(256, (3, 3), activation='relu', kernel_initializer='he_normal', padding='same')(p2)
    c3 = Conv2D(256, (3, 3), activation='relu', kernel_initializer='he_normal', padding='same')(c3)
    p3 = MaxPooling2D((2, 2))(c3)

    c4 = Conv2D(512, (3, 3), activation='relu', kernel_initializer='he_normal', padding='same')(p3)
    c4 = Conv2D(512, (3, 3), activation='relu', kernel_initializer='he_normal', padding='same')(c4)
    p4 = MaxPooling2D(pool_size=(2, 2))(c4)

    # Bridge
    c5 = Conv2D(1024, (3, 3), activation='relu', kernel_initializer='he_normal', padding='same')(p4)
    c5 = Conv2D(1024, (3, 3), activation='relu', kernel_initializer='he_normal', padding='same')(c5)

    # Decoder
    u6 = Conv2DTranspose(512, (2, 2), strides=(2, 2), padding='same')(c5)
    u6 = concatenate([u6, c4])
    c6 = Conv2D(512, (3, 3), activation='relu', kernel_initializer='he_normal', padding='same')(u6)
    c6 = Conv2D(512, (3, 3), activation='relu', kernel_initializer='he_normal', padding='same')(c6)

    u7 = Conv2DTranspose(256, (2, 2), strides=(2, 2), padding='same')(c6)
    u7 = concatenate([u7, c3])
    c7 = Conv2D(256, (3, 3), activation='relu', kernel_initializer='he_normal', padding='same')(u7)
    c7 = Conv2D(256, (3, 3), activation='relu', kernel_initializer='he_normal', padding='same')(c7)

    u8 = Conv2DTranspose(128, (2, 2), strides=(2, 2), padding='same')(c7)
    u8 = concatenate([u8, c2])
    c8 = Conv2D(128, (3, 3), activation='relu', kernel_initializer='he_normal', padding='same')(u8)
    c8 = Conv2D(128, (3, 3), activation='relu', kernel_initializer='he_normal', padding='same')(c8)

    u9 = Conv2DTranspose(64, (2, 2), strides=(2, 2), padding='same')(c8)
    u9 = concatenate([u9, c1], axis=3)
    c9 = Conv2D(64, (3, 3), activation='relu', kernel_initializer='he_normal', padding='same')(u9)
    c9 = Conv2D(64, (3, 3), activation='relu', kernel_initializer='he_normal', padding='same')(c9)

    outputs = Conv2D(num_classes, (1, 1), activation='softmax')(c9)

    model = Model(inputs=[inputs], outputs=[outputs])
    return model

# 6. Build EfficientNet-based Model
def build_efficientnet_segmentation(input_shape=(128, 128, 3), num_classes=12):
    # Use EfficientNetB0 as encoder
    base_model = EfficientNetB0(weights='imagenet', include_top=False, input_shape=input_shape)

    # Get skip connections
    skip_connection_names = ['block2a_expand_activation', 'block3a_expand_activation',
                           'block4a_expand_activation', 'block6a_expand_activation']
    skip_connections = [base_model.get_layer(name).output for name in skip_connection_names]

    # Decoder
    x = base_model.output

    # Upsampling path
    x = Conv2DTranspose(512, (2, 2), strides=(2, 2), padding='same')(x)
    x = concatenate([x, skip_connections[3]])
    x = Conv2D(512, (3, 3), activation='relu', padding='same')(x)

    x = Conv2DTranspose(256, (2, 2), strides=(2, 2), padding='same')(x)
    x = concatenate([x, skip_connections[2]])
    x = Conv2D(256, (3, 3), activation='relu', padding='same')(x)

    x = Conv2DTranspose(128, (2, 2), strides=(2, 2), padding='same')(x)
    x = concatenate([x, skip_connections[1]])
    x = Conv2D(128, (3, 3), activation='relu', padding='same')(x)

    x = Conv2DTranspose(64, (2, 2), strides=(2, 2), padding='same')(x)
    x = concatenate([x, skip_connections[0]])
    x = Conv2D(64, (3, 3), activation='relu', padding='same')(x)

    x = Conv2DTranspose(32, (2, 2), strides=(2, 2), padding='same')(x)
    x = Conv2D(32, (3, 3), activation='relu', padding='same')(x)

    outputs = Conv2D(num_classes, (1, 1), activation='softmax')(x)

    model = Model(inputs=base_model.input, outputs=outputs)
    return model

# 7. Training function
def train_model(model, X_train, y_train, X_test, y_test, optimizer, model_name, epochs=50):
    try:
        # Clear any previous sessions
        tf.keras.backend.clear_session()

        model.compile(optimizer=optimizer,
                     loss='categorical_crossentropy',
                     metrics=['accuracy'])

        callbacks = [
            EarlyStopping(patience=10, restore_best_weights=True, verbose=1),
            ReduceLROnPlateau(factor=0.5, patience=5, min_lr=1e-7, verbose=1)
        ]

        print(f"\nTraining {model_name} with {optimizer.__class__.__name__} optimizer...")
        print(f"Training data shape: {X_train.shape}, Labels shape: {y_train.shape}")

        # Convert to TensorFlow tensors if needed
        X_train = tf.convert_to_tensor(X_train, dtype=tf.float32)
        y_train = tf.convert_to_tensor(y_train, dtype=tf.float32)
        X_test = tf.convert_to_tensor(X_test, dtype=tf.float32)
        y_test = tf.convert_to_tensor(y_test, dtype=tf.float32)

        history = model.fit(X_train, y_train,
                           batch_size=8,  # Reduced batch size
                           epochs=epochs,
                           validation_data=(X_test, y_test),
                           callbacks=callbacks,
                           verbose=1)

        return history

    except Exception as e:
        print(f"Error during training {model_name}: {e}")
        return None

# 8. Evaluation function
def evaluate_model(model, X_test, y_test, model_name, optimizer_name):
    test_loss, test_acc = model.evaluate(X_test, y_test, verbose=0)
    print(f"{model_name} with {optimizer_name} - Test Loss: {test_loss:.4f}, Test Accuracy: {test_acc:.4f}")
    return test_loss, test_acc

# 9. Plotting function
def plot_training_history(histories, model_names, optimizer_names):
    fig, axes = plt.subplots(2, 2, figsize=(15, 10))

    # Plot training loss
    axes[0, 0].set_title('Training Loss')
    for i, (history, model_name, opt_name) in enumerate(zip(histories, model_names, optimizer_names)):
        axes[0, 0].plot(history.history['loss'], label=f'{model_name}-{opt_name}')
    axes[0, 0].set_xlabel('Epoch')
    axes[0, 0].set_ylabel('Loss')
    axes[0, 0].legend()
    axes[0, 0].grid(True)

    # Plot validation loss
    axes[0, 1].set_title('Validation Loss')
    for i, (history, model_name, opt_name) in enumerate(zip(histories, model_names, optimizer_names)):
        axes[0, 1].plot(history.history['val_loss'], label=f'{model_name}-{opt_name}')
    axes[0, 1].set_xlabel('Epoch')
    axes[0, 1].set_ylabel('Loss')
    axes[0, 1].legend()
    axes[0, 1].grid(True)

    # Plot training accuracy
    axes[1, 0].set_title('Training Accuracy')
    for i, (history, model_name, opt_name) in enumerate(zip(histories, model_names, optimizer_names)):
        axes[1, 0].plot(history.history['accuracy'], label=f'{model_name}-{opt_name}')
    axes[1, 0].set_xlabel('Epoch')
    axes[1, 0].set_ylabel('Accuracy')
    axes[1, 0].legend()
    axes[1, 0].grid(True)

    # Plot validation accuracy
    axes[1, 1].set_title('Validation Accuracy')
    for i, (history, model_name, opt_name) in enumerate(zip(histories, model_names, optimizer_names)):
        axes[1, 1].plot(history.history['val_accuracy'], label=f'{model_name}-{opt_name}')
    axes[1, 1].set_xlabel('Epoch')
    axes[1, 1].set_ylabel('Accuracy')
    axes[1, 1].legend()
    axes[1, 1].grid(True)

    plt.tight_layout()
    plt.show()

# Main execution
def main():
    try:
        # Load and preprocess data
        image_dir = '/content/SegNet-Tutorial/CamVid/train'
        mask_dir = '/content/SegNet-Tutorial/CamVid/trainannot'

        images, masks = load_dataset(image_dir, mask_dir)
        if images is None or masks is None:
            print("Failed to load dataset")
            return

        X_train, X_test, y_train, y_test = split_dataset(images, masks)
        if X_train is None:
            print("Failed to split dataset")
            return

        X_train, y_train = preprocess_data(X_train, y_train)
        X_test, y_test = preprocess_data(X_test, y_test)
        if X_train is None:
            print("Failed to preprocess data")
            return

        plot_samples(X_train, y_train, num_samples=3)

        # Define optimizers (choosing best ones)
        optimizers = [
            Adam(learning_rate=0.001),
            RMSprop(learning_rate=0.001)
        ]
        optimizer_names = ['Adam', 'RMSprop']

        # Training and evaluation
        histories = []
        results = []
        model_names = ['U-Net', 'EfficientNet']

        for model_name in model_names:
            for optimizer, opt_name in zip(optimizers, optimizer_names):
                print(f"\n{'='*50}")
                print(f"Training {model_name} with {opt_name}")
                print(f"{'='*50}")

                # Create a fresh model for each training
                try:
                    if model_name == 'U-Net':
                        current_model = build_unet()
                    else:
                        current_model = build_efficientnet_segmentation()

                    print(f"Model {model_name} created successfully")

                    # Train model
                    history = train_model(current_model, X_train, y_train, X_test, y_test,
                                        optimizer, model_name, epochs=50)

                    if history is None:
                        print(f"Training failed for {model_name} with {opt_name}")
                        continue

                    histories.append(history)

                    # Get training metrics
                    train_loss = min(history.history['loss'])
                    train_acc = max(history.history['accuracy'])
                    val_loss = min(history.history['val_loss'])
                    val_acc = max(history.history['val_accuracy'])

                    # Evaluate on test set
                    test_loss, test_acc = evaluate_model(current_model, X_test, y_test, model_name, opt_name)

                    # Store results
                    results.append({
                        'Model': model_name,
                        'Optimizer': opt_name,
                        'Train Loss': f"{train_loss:.4f}",
                        'Train Acc': f"{train_acc:.4f}",
                        'Val Loss': f"{val_loss:.4f}",
                        'Val Acc': f"{val_acc:.4f}",
                        'Test Loss': f"{test_loss:.4f}",
                        'Test Acc': f"{test_acc:.4f}"
                    })

                    # Clear memory
                    del current_model
                    tf.keras.backend.clear_session()

                except Exception as e:
                    print(f"Error with {model_name} and {opt_name}: {e}")
                    continue

        # Plot results if we have any histories
        if histories:
            plot_training_history(histories,
                                [name for name in model_names for _ in optimizer_names][:len(histories)],
                                (optimizer_names * len(model_names))[:len(histories)])

        # Display results table
        if results:
            results_df = pd.DataFrame(results)
            print("\n" + "="*80)
            print("FINAL RESULTS SUMMARY")
            print("="*80)
            print(results_df.to_string(index=False))
            return results_df
        else:
            print("No results to display")
            return None

    except Exception as e:
        print(f"Error in main execution: {e}")
        return None

# Run the complete pipeline
if __name__ == "__main__":
    results = main()
```
